## What the dose looks like

Each training example pairs a Sanskrit sentence with its linearized graph, where
typed relations (PRAKARATA for qualifier–qualificand, VISAYATA for the
object-of-cognition relation, AVACCHEDAKA for scope) connect ontology nodes drawn
from the sapta-padārtha categories. Arm D trains on the linearized strings; the
graph and parse are retained as gold structure for analysis. The batch exporter
and the exact frozen corpus hashes live in the repository's corpus manifest.
